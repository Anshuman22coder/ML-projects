```python
# importing the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#data collection and procesing
#loading the data from csv file
car_dataset = pd.read_csv('/content/car data.csv')
car_dataset.head()
```

```python
car_dataset.shape
```

```python
car_dataset.isnull().sum()
```

```python
#categorical data ::text data like petrol or diesla in fuel type column
#cheking the distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

```python
# encodingg categorcial data ___label or one hot encoding,, pre-processing data
car_dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
car_dataset.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
car_dataset.replace({'Transmission':{'Manual':0,'Automatic':1,}},inplace=True)

car_dataset.head()
```

```python
#splitting to the training data
X=car_dataset.drop(["Car_Name","Present_Price"],axis=1)
Y=car_dataset["Present_Price"]
X.shape
print(Y)
```

```python
#Splitting to traiing and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
#model traiing
#load linear regression model..
linear_reg_model=LinearRegression()
linear_reg_model.fit(X_train,Y_train)
```

```python
#evalaution of model
predict=linear_reg_model.predict(X_train)
print(predict)
#R-squared error is used. not accuracy score bcznm,, we cannot use accuracy score in regression..
error_score=metrics.r2_score(Y_train,predict)
print("R squared error : ",error_score)  # as <<<1 then the model is good enough
```

```python
# making a predictive system
input_data = (2015,4.00,60000,0,0,0,0)  # the present price is 5.90 lakhs for this data
#changing input data to numpy
input_data_as_numpy_array = np.array(input_data, dtype=np.float64).reshape(1, -1)
#reshape the array as we are predicting for one instance
#input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)  # telling the model we are giving just one row ,,
print(input_data_as_numpy_array)
```

```python
predict_on_unknown_data=linear_reg_model.predict(input_data_as_numpy_array)
print(predict_on_unknown_data)
```

```python
#model evaluation on test data
Y_pred_test = linear_reg_model.predict(X_test)
test_error_score = metrics.r2_score(Y_test, Y_pred_test)
print("Test R squared error:{:.3f}".format(test_error_score*100))
```
